```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from typing import Tuple, Dict, List
import copy
np.random.seed(42)
```

```python
class GridWorldMDP:
    """
    A 5x5 Grid World MDP for the Delivery Robot problem.

    States: (row, col) positions in the grid
    Actions: UP, DOWN, LEFT, RIGHT
    Transitions: Stochastic with 0.8 probability of intended action
    """

    def __init__(self,
                 grid_size: int = 5,
                 goal_reward: float = 10.0,
                 crash_reward: float = -10.0,
                 living_reward: float = -0.04,
                 gamma: float = 0.9):
        """
        Initialize the Grid World MDP.

        Args:
            grid_size: Size of the square grid (default: 5)
            goal_reward: Reward for reaching the goal state (default: 10.0)
            crash_reward: Reward for reaching the crash state (default: -10.0)
            living_reward: Small negative reward for each step (default: -0.04)
            gamma: Discount factor (default: 0.9)
        """
        self.grid_size = grid_size
        self.goal_reward = goal_reward
        self.crash_reward = crash_reward
        self.living_reward = living_reward
        self.gamma = gamma

        # Define special states
        self.start_state = (4, 0)  # Bottom-left corner
        self.goal_state = (0, 4)   # Top-right corner
        self.crash_state = (4, 4)  # Bottom-right corner

        # Define terminal states
        self.terminal_states = [self.goal_state, self.crash_state]

        # Define actions
        self.actions = ['UP', 'DOWN', 'LEFT', 'RIGHT']

        # Define transition probabilities
        self.transition_prob = {
            'intended': 0.8,
            'perpendicular': 0.1  # 0.1 for each perpendicular direction
        }

        # Initialize all states
        self.states = [(i, j) for i in range(grid_size) for j in range(grid_size)]

        # Initialize value function
        self.values = {state: 0.0 for state in self.states}

        # Initialize policy
        self.policy = {state: None for state in self.states}
```

```python
def get_next_state(current_state: Tuple[int, int],
                   action: str,
                   grid_size: int) -> Tuple[int, int]:
    """
    Get the next state given current state and action.
    Returns the same state if action would go out of bounds.

    Args:
        current_state: Current (row, col) position
        action: Action to take ('UP', 'DOWN', 'LEFT', 'RIGHT')
        grid_size: Size of the grid

    Returns:
        Next state (row, col)
    """
    row, col = current_state

    if action == 'UP':
        new_row = max(0, row - 1)
        return (new_row, col)
    elif action == 'DOWN':
        new_row = min(grid_size - 1, row + 1)
        return (new_row, col)
    elif action == 'LEFT':
        new_col = max(0, col - 1)
        return (row, new_col)
    elif action == 'RIGHT':
        new_col = min(grid_size - 1, col + 1)
        return (row, new_col)

    return current_state

def get_transition_probabilities(mdp: GridWorldMDP,
                                 state: Tuple[int, int],
                                 action: str) -> Dict[Tuple[int, int], float]:
    """
    Get transition probabilities for taking an action from a state.

    Args:
        mdp: The GridWorld MDP instance
        state: Current state
        action: Action to take

    Returns:
        Dictionary mapping next states to their probabilities
    """
    if state in mdp.terminal_states:
        return {state: 1.0}

    transitions = {}

    # Intended action with 0.8 probability
    intended_next = get_next_state(state, action, mdp.grid_size)
    transitions[intended_next] = mdp.transition_prob['intended']

    # Perpendicular actions with 0.1 probability each
    if action in ['UP', 'DOWN']:
        perpendicular_actions = ['LEFT', 'RIGHT']
    else:  # action in ['LEFT', 'RIGHT']
        perpendicular_actions = ['UP', 'DOWN']

    for perp_action in perpendicular_actions:
        perp_next = get_next_state(state, perp_action, mdp.grid_size)
        if perp_next in transitions:
            transitions[perp_next] += mdp.transition_prob['perpendicular']
        else:
            transitions[perp_next] = mdp.transition_prob['perpendicular']

    return transitions
```

```python
def get_reward(mdp: GridWorldMDP, state: Tuple[int, int]) -> float:
    """
    Get the reward for being in a state.

    Args:
        mdp: The GridWorld MDP instance
        state: Current state

    Returns:
        Reward value
    """
    if state == mdp.goal_state:
        return mdp.goal_reward
    elif state == mdp.crash_state:
        return mdp.crash_reward
    else:
        return mdp.living_reward
```

```python
def value_iteration(mdp: GridWorldMDP,
                    epsilon: float = 0.001,
                    max_iterations: int = 1000) -> Tuple[Dict, List[float]]:
    """
    Perform value iteration to compute optimal state values.

    Args:
        mdp: The GridWorld MDP instance
        epsilon: Convergence threshold
        max_iterations: Maximum number of iterations

    Returns:
        Tuple of (final values dictionary, list of max value changes per iteration)
    """
    values = {state: 0.0 for state in mdp.states}
    value_changes = []

    for iteration in range(max_iterations):
        new_values = copy.deepcopy(values)
        max_change = 0

        for state in mdp.states:
            if state in mdp.terminal_states:
                new_values[state] = get_reward(mdp, state)
            else:
                # Compute Q-values for all actions
                q_values = []

                for action in mdp.actions:
                    q_value = 0
                    transitions = get_transition_probabilities(mdp, state, action)

                    for next_state, prob in transitions.items():
                        reward = get_reward(mdp, state)
                        q_value += prob * (reward + mdp.gamma * values[next_state])

                    q_values.append(q_value)

                # Update value with maximum Q-value
                new_values[state] = max(q_values)

            # Track maximum change
            change = abs(new_values[state] - values[state])
            max_change = max(max_change, change)

        values = new_values
        value_changes.append(max_change)

        # Check for convergence
        if max_change < epsilon:
            print(f"Value Iteration converged after {iteration + 1} iterations")
            break

    return values, value_changes
```

```python
def extract_policy(mdp: GridWorldMDP, values: Dict) -> Dict:
    """
    Extract the optimal policy from the computed value function.

    Args:
        mdp: The GridWorld MDP instance
        values: Computed value function

    Returns:
        Dictionary mapping states to optimal actions
    """
    policy = {}

    for state in mdp.states:
        if state in mdp.terminal_states:
            policy[state] = None
        else:
            # Compute Q-values for all actions
            q_values = {}

            for action in mdp.actions:
                q_value = 0
                transitions = get_transition_probabilities(mdp, state, action)

                for next_state, prob in transitions.items():
                    reward = get_reward(mdp, state)
                    q_value += prob * (reward + mdp.gamma * values[next_state])

                q_values[action] = q_value

            # Select action with maximum Q-value
            policy[state] = max(q_values.keys(), key=lambda k: q_values[k])

    return policy
```

```python
def visualize_grid_world(mdp: GridWorldMDP, values: Dict = None, policy: Dict = None):
    """
    Visualize the grid world with values and/or policy.

    Args:
        mdp: The GridWorld MDP instance
        values: Optional value function to display
        policy: Optional policy to display
    """
    fig, ax = plt.subplots(figsize=(10, 10))

    # Draw grid
    for i in range(mdp.grid_size + 1):
        ax.axhline(i, color='black', linewidth=1)
        ax.axvline(i, color='black', linewidth=1)

    # Action arrows for visualization
    action_arrows = {
        'UP': '↑',
        'DOWN': '↓',
        'LEFT': '←',
        'RIGHT': '→'
    }

    # Fill in states
    for state in mdp.states:
        row, col = state

        # Determine color based on state type
        if state == mdp.start_state:
            color = 'lightgreen'
            label = 'S'
        elif state == mdp.goal_state:
            color = 'gold'
            label = 'G'
        elif state == mdp.crash_state:
            color = 'lightcoral'
            label = 'X'
        else:
            color = 'white'
            label = ''

        # Add colored rectangle
        rect = mpatches.Rectangle((col, mdp.grid_size - row - 1),
                                  1, 1, facecolor=color, edgecolor='black')
        ax.add_patch(rect)

        # Add state label
        if label:
            ax.text(col + 0.5, mdp.grid_size - row - 0.5, label,
                   ha='center', va='center', fontsize=20, fontweight='bold')

        # Add value if provided
        if values and state not in mdp.terminal_states:
            ax.text(col + 0.5, mdp.grid_size - row - 0.7,
                   f'{values[state]:.2f}',
                   ha='center', va='center', fontsize=10)

        # Add policy arrow if provided
        if policy and policy[state]:
            ax.text(col + 0.5, mdp.grid_size - row - 0.3,
                   action_arrows[policy[state]],
                   ha='center', va='center', fontsize=16)

    ax.set_xlim(0, mdp.grid_size)
    ax.set_ylim(0, mdp.grid_size)
    ax.set_aspect('equal')
    ax.set_xticks(range(mdp.grid_size + 1))
    ax.set_yticks(range(mdp.grid_size + 1))
    ax.set_title('Delivery Robot Grid World MDP', fontsize=16, fontweight='bold')

    # Add legend
    legend_elements = [
        mpatches.Patch(color='lightgreen', label='Start (S)'),
        mpatches.Patch(color='gold', label='Goal (G) +10'),
        mpatches.Patch(color='lightcoral', label='Crash (X) -10'),
        mpatches.Patch(color='white', label='Normal State')
    ]
    ax.legend(handles=legend_elements, loc='upper center',
             bbox_to_anchor=(0.5, -0.05), ncol=4)

    plt.show()

def plot_convergence(value_changes: List[float]):
    """
    Plot the convergence of value iteration.

    Args:
        value_changes: List of maximum value changes per iteration
    """
    plt.figure(figsize=(10, 6))
    plt.plot(value_changes, 'b-', linewidth=2)
    plt.xlabel('Iteration', fontsize=12)
    plt.ylabel('Maximum Value Change', fontsize=12)
    plt.title('Value Iteration Convergence', fontsize=14, fontweight='bold')
    plt.yscale('log')
    plt.grid(True, alpha=0.3)
    plt.show()
```

```python
print("=" * 60)
print("Running MDP with Default Parameters")
print("=" * 60)

mdp_default = GridWorldMDP(
    grid_size=5,
    goal_reward=10.0,
    crash_reward=-10.0,
    living_reward=-0.04,
    gamma=0.9
)

print(f"Grid Size: {mdp_default.grid_size}x{mdp_default.grid_size}")
print(f"Goal Reward: {mdp_default.goal_reward}")
print(f"Crash Reward: {mdp_default.crash_reward}")
print(f"Living Reward: {mdp_default.living_reward}")
print(f"Discount Factor (γ): {mdp_default.gamma}")
print()

# Run value iteration
values_default, changes_default = value_iteration(mdp_default)

# Extract optimal policy
policy_default = extract_policy(mdp_default, values_default)

# Visualize results
print("\nGrid World with Optimal Values and Policy:")
visualize_grid_world(mdp_default, values_default, policy_default)

print("\nConvergence Plot:")
plot_convergence(changes_default)

# Print optimal path from start to goal
print("\nOptimal Path from Start to Goal:")
current = mdp_default.start_state
path = [current]
steps = 0
max_steps = 20  # Prevent infinite loops

while current != mdp_default.goal_state and steps < max_steps:
    action = policy_default[current]
    if action is None:
        break
    # Get most likely next state (intended action)
    current = get_next_state(current, action, mdp_default.grid_size)
    path.append(current)
    steps += 1

print(f"Path: {' -> '.join([str(s) for s in path])}")
```

```python
print("\n" + "=" * 60)
print("Experiment 1: Effect of Discount Factor (γ)")
print("=" * 60)

gamma_values = [0.1, 0.5, 0.9, 0.99]
fig, axes = plt.subplots(2, 2, figsize=(15, 15))
axes = axes.flatten()

for idx, gamma in enumerate(gamma_values):
    print(f"\nTesting γ = {gamma}")

    mdp = GridWorldMDP(
        grid_size=5,
        goal_reward=10.0,
        crash_reward=-10.0,
        living_reward=-0.04,
        gamma=gamma
    )

    values, changes = value_iteration(mdp)
    policy = extract_policy(mdp, values)

    # Visualize in subplot
    ax = axes[idx]

    # Draw grid
    for i in range(mdp.grid_size + 1):
        ax.axhline(i, color='black', linewidth=1)
        ax.axvline(i, color='black', linewidth=1)

    action_arrows = {'UP': '↑', 'DOWN': '↓', 'LEFT': '←', 'RIGHT': '→'}

    for state in mdp.states:
        row, col = state

        if state == mdp.start_state:
            color = 'lightgreen'
        elif state == mdp.goal_state:
            color = 'gold'
        elif state == mdp.crash_state:
            color = 'lightcoral'
        else:
            color = 'white'

        rect = mpatches.Rectangle((col, mdp.grid_size - row - 1),
                                  1, 1, facecolor=color, edgecolor='black')
        ax.add_patch(rect)

        if policy[state]:
            ax.text(col + 0.5, mdp.grid_size - row - 0.5,
                   action_arrows[policy[state]],
                   ha='center', va='center', fontsize=14)

    ax.set_xlim(0, mdp.grid_size)
    ax.set_ylim(0, mdp.grid_size)
    ax.set_aspect('equal')
    ax.set_title(f'γ = {gamma}', fontsize=14, fontweight='bold')
    ax.set_xticks([])
    ax.set_yticks([])

plt.suptitle('Optimal Policies for Different Discount Factors',
             fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
